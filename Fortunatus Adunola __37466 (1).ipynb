```python
# Create the years and durations lists
years = list(range(2011, 2021))
durations = [103, 101, 99, 100, 100, 95, 95, 96, 93, 90]
```

```python
# Create a dictionary with the two lists
movie_dict = {
    "years": years,
    "durations": durations
}
```

```python
# Print the dictionary
print(movie_dict)
```

```python
# Import pandas under its usual alias
import pandas as pd
```

```python
# Create a DataFrame
durations_df = pd.DataFrame(movie_dict)
```

```python
# Print the DataFrame
print(durations_df)
```

```python
# Import matplotlib.pyplot under its usual alias and create a figure
import matplotlib.pyplot as plt
```

```python
# Create a new figure
plt.figure()
```

```python
# Create a line plot
plt.plot(durations_df['years'], durations_df['durations'], marker='o')
plt.title("Netflix Movie Durations 2011-2020")
plt.xlabel("Year")
plt.ylabel("Average Movie Duration (minutes)")
```

```python
#Read in the CSV as a Data Frame
csv_path = "downloads/netflix_data.csv"
netflix_df = pd.read_csv(csv_path)
```

```python
# Print the first five rows of the DataFrame
print(netflix_df.head())
```

```python
# Subset the DataFrame for type "Movie"
netflix_df_movies_only = netflix_df[netflix_df['type'] == 'Movie']
```

```python
# Select only the columns of interest
netflix_movies_col_subset = ['title', 'country', 'genre', 'release_year', 'duration']
netflix_movies_col_subset = netflix_df_movies_only
```

```python
# Print the first five rows of the new DataFrame
print(netflix_movies_col_subset.head())
```

```python
# Create a figure and increase the figure size
plt.figure(figsize=(12,8))
```

```python
# Create a scatter plot of duration versus year
plt.scatter(netflix_movies_col_subset['release_year'], netflix_movies_col_subset['duration'], alpha=0.5)
plt.title("Movie Duration by Year of Release")
plt.xlabel("Release Year")
plt.ylabel("Movie Duration (minutes)")
```

```python
# Show the plot
plt.show()
```

```python
# Filter for durations shorter than 60 minutes
short_movies = netflix_movies_col_subset[netflix_movies_col_subset['duration'] < 60]
```

```python
# Print the first 20 rows of short_movies
print(short_movies.head(20))
```

```python
# Define an empty list
colors = []
```

```python
# Iterate over rows of netflix_movies_col_subset
for index, row in netflix_movies_col_subset.iterrows():
    if row['genre'] == 'Children':
        colors.append('red')
    elif row['genre'] == 'Documentaries':
        colors.append('blue')
    elif row['genre'] == 'Stand-Up':
        colors.append('green')
    else:
        colors.append('black')
```

```python
# Inspect the first 10 values in your list
print(colors[:10])
```

```python
# Set the figure style and initalize a new figure
plt.style.use('fivethirtyeight')
plt.figure(figsize=(12,8))
```

```python
# Create a scatter plot of duration versus release_year, and add label
plt.scatter(netflix_movies_col_subset['release_year'], netflix_movies_col_subset['duration'], c=colors, alpha=0.5)
plt.title("Movie Duration by Year of Release")
plt.xlabel("Release Year")
plt.ylabel("Duration (min)")
```

```python
# Show the plot
plt.show()
```
